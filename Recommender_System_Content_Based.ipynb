```python
movies=df[['id','title','overview','genre']]
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genre']
```

```python
movies.head()
```

```python
movies.info()
```

```python
data=movies.drop(columns=['overview','genre'])
data.head()
```

```python
# //to convert text into vectors,
# 1. Bag of word
# 2. TFIDF
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector=cv.fit_transform(data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
# we will use cosine similarity now
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
data[data['title']=="The Godfather"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key= lambda v:v[1])
for i in distance[0:5]:
  print(data.iloc[i[0]].title)
```

```python
# enumerate-> it pairs each similarity score with its index
```

```python
def recommend(movies):
  index=data[data['title']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key= lambda v:v[1])
  for i in distance [0:5]:
    print(data.iloc[i[0]].title)

recommend('Avatar')
```

```python
recommend('Iron Man')
```

# Task
This project successfully developed a content-based movie recommendation system.

The steps taken were as follows:

1.  **Data Loading and Initial Exploration**: The `movieDataseet.csv` dataset was loaded into a pandas DataFrame. Initial checks were performed using `df.isnull().any()`, `df.describe()`, `df.info()`, and `df.isnull().sum()` to understand the data structure, identify missing values (in 'genre' and 'overview'), and inspect summary statistics.
2.  **Feature Selection**: Relevant columns (`id`, `title`, `overview`, `genre`) were selected for building the recommendation engine.
3.  **Feature Engineering**: A new column named 'tags' was created by concatenating the 'overview' and 'genre' columns. This combined text was intended to represent the core content of each movie.
4.  **Data Cleaning**: The original 'overview' and 'genre' columns were dropped, keeping only 'id', 'title', and 'tags' for the final recommendation data.
5.  **Text Vectorization**: The 'tags' content was converted into numerical vectors using `CountVectorizer`. This technique transforms text into a matrix of token counts, with a maximum of 10,000 features and common English stop words removed.
6.  **Similarity Calculation**: Cosine similarity was calculated between all movie vectors. This metric measures the cosine of the angle between two vectors, indicating their similarity.
7.  **Recommendation Function**: A Python function `recommend(movies)` was created. This function takes a movie title as input, finds its index, calculates its similarity to all other movies, sorts them by similarity, and prints the top 5 most similar movie titles.

The goal achieved is a functional content-based movie recommendation system that can suggest movies similar to a given input based on their plot overview and genres.
